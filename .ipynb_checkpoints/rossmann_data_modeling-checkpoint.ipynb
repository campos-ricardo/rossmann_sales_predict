```python
# Function responsible to calculate de cramer_v coeficient between two variables
def cramer_v( x, y ):
    
    cm = pd.crosstab( x, y ).to_numpy()
    n = cm.sum()
    r, k = cm.shape
    chi2 = stats.chi2_contingency( cm )[0]
    chi2corr = max( 0, chi2 - (k-1)*(r-1)/(n-1) )
    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)
    
    return np.sqrt( (chi2corr/n) / ( min( kcorr-1, rcorr-1 ) ) )
```

```python
# Function responsible for calculating the cramer_v coeficient of a table fo variables and then retunr a data frame
def cramer_v_table (df):
    
    cols_1 = list(df.columns)
    cols_2 = list(df.columns)
    
    cramer_table = pd.DataFrame(columns = cols_1, index = cols_1)
    
    for columns_1 in cols_1:
        for columns_2 in cols_2:
            
            cramer_table.loc[columns_2,columns_1] =  cramer_v(df[columns_1],df[columns_2])
        
    return cramer_table
```

```python
# Calculates three diffrent types of error : MAE, MAPE annd RMSE. And then returns a dataframe
def model_performance (model_name,yhat,y):
    
    mae = mean_absolute_error(y,yhat)
    mape = mean_absolute_percentage_error(y,yhat)
    rmse = np.sqrt(mean_squared_error(y,yhat))
    
    performance_dict = {'model_name':model_name,
                        'mae':mae,
                        'mape':mape,
                        'rmse':rmse}
    
    model_metrics = pd.DataFrame(performance_dict, index = [0])
    return model_metrics
```

```python
# Function to evaluate a XGBoost model 
def xgb_cv(n_estimators, eta, max_depth, subsample, data, targets):
    
    estimator = xgb.XGBRegressor(
        objective='reg:squarederror',
        n_estimators=n_estimators,
        max_depth=max_depth,
        eta=eta,
        subsample=subsample,
    )
    cval = cross_val_score(estimator, data, targets,
                           scoring='neg_mean_absolute_error', cv=4)
    return cval.mean()
```

```python
def optimize_xgb(data, targets):
    """Apply Bayesian Optimization to XGBoost parameters."""
    def xgb_crossval(n_estimators, eta, max_depth, subsample):
        """Wrapper of XGBoost cross validation. """
        return xgb_cv(
            n_estimators=int(n_estimators),
            eta = float(eta),
            max_depth=int(max_depth),
            #subsample=max(min(max_features, 0.999), 1e-3),
            subsample = float(subsample),
            data=data,
            targets=targets,
        )

    optimizer = BayesianOptimization(xgb_crossval,{"n_estimators":(10,250),"eta":(0,1),"max_depth":(3,10),
                                                  "subsample":(0,1)},verbose=2)
    
    optimizer.maximize(n_iter=10)

    print("Final result:", optimizer.max)
```

### 4.2 - Encoding

#### 4.2.1 - Dummie and Label Encoding

```python
 # state_holiday - One Hot Encoding
df_4 = pd.get_dummies( df_4, prefix=['state_holiday'], columns=['state_holiday'] )
```

```python
# store_type - Label Encoding
le = LabelEncoder()
df_4['store_type'] = le.fit_transform( df_4['store_type'] )
pickle.dump( le, open( r'C:\Users\Pichau\OneDrive\Documents\repos\ds_em_producao\rossmann_predict_deploy\parameters\/store_type_label_encoder.pkl', 'wb') )

# assortment - Ordinal Encoding - utilizando um dicionário
assortment_dict = {'basic': 1, 'extra': 2, 'extended': 3}
df_4['assortment'] = df_4['assortment'].map( assortment_dict )
```

#### 4.2.2 - Response Variable Enconding

```python
df_4['sales'] = np.log1p(df_4['sales'])
```

#### 4.2.3 - Circular Encoding

```python
# Todas as variáveis circulares como dia e mês podem ser convertidas em senos e cossenos com os angulos definidos a partir da teoria do arco de circuferencia
# Um arco completo possui 360 graus ou 2*pi radianos
# Utilizando a regra de três é possível deduzir o angulo a partir do comprimento de arco utilizado  alpha = (2*pi/n)*(comprimento de arco)
# n - número de divisoes dentro do circulo de 360

# Mes
df_4['competition_open_since_month_sin'] = df_4['competition_open_since_month'].apply(lambda x: np.sin((2*np.pi/12)*x))
df_4['competition_open_since_month_cos'] = df_4['competition_open_since_month'].apply(lambda x: np.cos((2*np.pi/12)*x))
df_4['month_sin'] = df_4['month'].apply(lambda x: np.sin((2*np.pi/12)*x))
df_4['month_cos'] = df_4['month'].apply(lambda x: np.cos((2*np.pi/12)*x))

# Dia da semana
df_4['day_of_week_sin'] = df_4['day_of_week'].apply(lambda x: np.sin((2*np.pi/7)*x))
df_4['day_of_week_cos'] = df_4['day_of_week'].apply(lambda x: np.cos((2*np.pi/7)*x))

# Semana do ano 
df_4['promo2_since_week_sin'] = df_4['promo2_since_week'].apply(lambda x: np.sin((2*np.pi/52)*x))
df_4['promo2_since_week_cos'] = df_4['promo2_since_week'].apply(lambda x: np.cos((2*np.pi/52)*x))
df_4['week_of_year_sin'] = df_4['weekofyear'].apply(lambda x: np.sin((2*np.pi/52)*x))
df_4['week_of_year_cos'] = df_4['weekofyear'].apply(lambda x: np.cos((2*np.pi/52)*x))

# Dia
df_4['day_sin'] = df_4['day'].apply(lambda x: np.sin((2*np.pi/31)*x))
df_4['day_cos'] = df_4['day'].apply(lambda x: np.cos((2*np.pi/31)*x))
```

```python
# Dropando colunas que foram convertidas
df_4.drop(columns = ['day','weekofyear','promo2_since_week','day_of_week','month','competition_open_since_month','competition_since','promo_since'],inplace = True)
```
